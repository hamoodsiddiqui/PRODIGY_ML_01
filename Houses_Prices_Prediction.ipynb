# Removing Outliers

```python
def outlier_remove(df, *args):
    prep_df =df.copy()
    for col in args:
        q3 = prep_df[col].quantile(0.75)
        q1 = prep_df[col].quantile(0.25)
        iqr = q3-q1
        upper_fence = q3 + 1.5 * iqr
        lower_fence = q1 - 1.5 * iqr
        prep_df = prep_df [~((prep_df [col]>upper_fence) | (prep_df [col]<lower_fence))]
    return prep_df
```

# Splitting Data into training and testing

```python
X = training_data.drop("SalePrice", axis = 1)
y = training_data["SalePrice"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
lr_model = LinearRegression()

lr_model.fit(X_train, y_train)
```

```python
predictions = lr_model.predict(X_test)
```

```python
predictions
```

```python
y_test
```

```python
for prediction, actual_value in zip(predictions, y_test):
    print(prediction, actual_value)
```

```python
r2_score(y_test, predictions) * 100
```

```python
mean_squared_error(y_test, predictions)
```
